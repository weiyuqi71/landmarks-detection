```python
import glob
import os
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms,models
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
from PIL import Image
import matplotlib.image as mpimg

from homo_util import hartley_normalization
from homo_util import feat2HomoConstraints
from homo_util import estHomography


from mtcnn import MTCNN
device = "cpu"
    
def image_transform(imagepath):
    resize_size =256
    output_size = 224
    image = cv2.imread(imagepath)
    
    if(image.shape[2] == 4):
            image = image[:,:,0:3]
            
    h = image.shape[0]
    w = image.shape[1]

    if h < w:
        top, bottom = int((w-h)/2), int((w-h)/2)
        left, right = 0,0
        color = [0,0,0]
        image = cv2.copyMakeBorder(image, top, bottom, left, right, cv2.BORDER_CONSTANT,value=color)

    elif h > w:
        top, bottom = 0,0
        left, right = int((h-w)/2), int((h-w)/2)
        color = [0, 0, 0]
        image = cv2.copyMakeBorder(image, top, bottom, left, right, cv2.BORDER_CONSTANT,value=color)
            
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    gray_image = gray_image / 255.0 
    gray_image = cv2.resize(gray_image, (resize_size, resize_size))
    start = int((resize_size-output_size)/2)
    end = int(resize_size-(resize_size-output_size)/2)
    gray_image = gray_image[start:end,start:end]
    image_tensor=torch.from_numpy(gray_image.reshape(1,1,224,224)) 
    image_tensor = image_tensor.type(torch.FloatTensor)
    
    image = cv2.resize(image, (resize_size, resize_size))
    image = image[start:end,start:end]
    
    return image, image_tensor
```

```python
def prediction(weight_path,imagepath):
    
    net = torch.load(weight_path, map_location=torch.device('cpu'))
    net.eval()
    image, image_tensor = image_transform(imagepath)
    image_tensor=image_tensor.to(device)
    output_pts = net(image_tensor)

    torch.squeeze(output_pts) 
    output_pts = output_pts.view(4, -1)
    predicted_key_pts = output_pts.data.cpu().numpy()
    predicted_key_pts = predicted_key_pts*50+100
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.scatter(predicted_key_pts[:, 0], predicted_key_pts[:, 1], s=60, marker='.', c='m')
    plt.axis("off")
    plt.show()
    return image,predicted_key_pts
```

```python
weight_path = "weights.pth"
imagepath = "1587024198015.jpg"
image,predicted_key_pts = prediction(weight_path,imagepath)
```

```python
def idcrop(imagepath):
    image,predicted_key_pts = prediction("weights.pth",imagepath)
    crop = np.array([[predicted_key_pts[0][0],predicted_key_pts[1][0],predicted_key_pts[2][0],predicted_key_pts[3][0]],
                        [predicted_key_pts[0][1],predicted_key_pts[1][1],predicted_key_pts[2][1],predicted_key_pts[3][1]], 
                        [1, 1, 1, 1]])

    shape = np.array([[0, 160, 160, 0],
                        [0, 0, 100, 100], 
                        [1, 1, 1, 1]])

    feat1_nh, T1 = hartley_normalization(crop.T)
    feat2_nh, T2 = hartley_normalization(shape.T)


    A_hart = feat2HomoConstraints(feat1_nh, feat2_nh)

    H_w = estHomography(A_hart)
    H_w = np.array(np.linalg.inv(T2)@H_w@(T1))

    im_out = cv2.warpPerspective(image, H_w, (160, 100))
    plt.axis("off")
    plt.imshow(im_out)
idcrop(imagepath)
```
